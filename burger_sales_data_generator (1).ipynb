# **â² Time duration to generate the dataset**


```python
# Define the time range for one year
start_date = datetime(2024, 1, 1, 9, 0, 0)  # Start date and time
end_date = datetime(2024, 12, 31, 22, 30, 0)  # End date and time
```

## **ðŸ—» Pick hours for a restaurant**

```python
# Define the time range for peak hours
peak_start_time = datetime.strptime('16:00:00', '%H:%M:%S').time()  # Start of peak hours
peak_end_time = datetime.strptime('20:30:00', '%H:%M:%S').time()  # End of peak hours
```

# **ðŸ”€ Generate the dataset**

```python
# Generate random sales data for each burger type at each time interval
sales_data = []
for time in pd.date_range(start=start_date, end=end_date, freq='15min'):
    for burger, ingredients in burgers.items():
        # Determine the sales multiplier based on time, month, and burger type
        if time.month in [1, 2, 3, 12]:  # Winter months
            sales_multiplier = random.random() * 0.5 + 0.5  # Decreased sales
        elif time.month in [6, 7, 8]:  # Summer months
            sales_multiplier = random.random() * 1.0 + 2.5  # Increased sales
        elif time.time() >= peak_start_time and time.time() <= peak_end_time:
            sales_multiplier = random.random() * 1.0 + 1.5  # Increase sales during peak hours
        else:
            sales_multiplier = 1.0  # Regular sales outside peak hours

        # Adjust sales multiplier based on burger characteristics
        if 'Small Bun' in ingredients:
            sales_multiplier *= random.random() * 1.7 + 2.0  # Add variation for small size burgers
        elif 'Large Bun' in ingredients:
            sales_multiplier *= random.random() * 1.95 + 0.5  # Decrease sales multiplier for burgers with large buns
        elif 'Whopper' in burger:
            sales_multiplier *= 2.3  # Medium sales for Whopper burger
        elif 'Long Chicken' in burger:
            sales_multiplier *= random.random() * 1.0 + 2.0  # Medium sales for Long Chicken burger
        elif 'Cheese' in burger:
            sales_multiplier *= random.random() * 2.4 + 2.0  # Medium sales for Cheese burger
        elif 'Plant' in burger or 'Veggi King' in burger:
            sales_multiplier *= random.random() * 0.5 + 2.0  # Medium sales for Plant-based burgers

        # Generate random quantity sold and calculate ingredient consumption and total price
        quantity_sold = min(int(random.random() * 5 * sales_multiplier), 5)  # Cap at 5
        ingredient_consumption = {ingredient: ingredients[ingredient] * quantity_sold for ingredient in ingredients}
        total_price = quantity_sold * burger_prices[burger]

        # Determine base type (Plant-Based or Beef-Based)
        base_type = 'Plant-Based' if 'Plant' in burger or 'Veggi King' in burger else 'Beef-Based'

        # Append sales data for the current time and burger type
        sales_data.append({'Time': time, 'Burger Name': burger, 'Base type': base_type, 'Quantity Sold': quantity_sold, 'Price': total_price, **ingredient_consumption})
```

# **The generated dataset**

```python
# Create a DataFrame from the sales data
df = pd.DataFrame(sales_data)
# Export DataFrame to a CSV file
df = df.fillna(0)  # Replace NaN values with 0 for ingredient consumption
df.head()
```

## **Generated csv file**

```python
df.to_csv('Burger sales data_one_year.csv', index=False)
```

# **Total ðŸ’°turnover report**

```python
# Function to generate report with burgersold

def sales_report_by_burger_visualization(start_date, end_date, months, start_time=None, end_time=None):
    # Filter sales data based on provided dates and months
    filtered_df = df[(df['Time'] >= start_date) & (df['Time'] <= end_date) & (df['Time'].dt.month.isin(months))]

    # Further filter data based on provided time range, if given
    if start_time and end_time:
        filtered_df = filtered_df[(filtered_df['Time'].dt.time >= start_time) & (filtered_df['Time'].dt.time <= end_time)]

    # Group filtered data by burger name and calculate total quantity sold and total turnover
    report = filtered_df.groupby('Burger Name').agg({'Quantity Sold': 'sum', 'Price': 'sum'}).reset_index()
    report.rename(columns={'Quantity Sold': 'Total Quantity Sold', 'Price': 'Turnover'}, inplace=True)

    # Create a bar chart visualization for the sales report
    fig = px.bar(report, x='Burger Name', y='Turnover', color='Burger Name', title='Sales Report by Burger',
                 labels={'Turnover': 'Turnover ($)'}, text='Turnover')

    # Customize the appearance of the bar chart
    fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
    fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')

    return fig
```

## **ðŸ›  Filters to adjust desired range**

```python
# Example usage with time filter
start_date = datetime(2024, 1, 1)  # Start date for the report
end_date = datetime(2024, 12, 31)  # End date for the report
months = [1, 2, 3, 6, 7, 8]  # Selected months: January, February, March, June, July, August

# Define the time range for filtering
start_time = datetime.strptime('09:00:00', '%H:%M:%S').time()  # Start time for the report
end_time = datetime.strptime('18:00:00', '%H:%M:%S').time()  # End time for the report

# Generate the sales report visualization for the specified date and time range
sales_report_fig = sales_report_by_burger_visualization(start_date, end_date, months, start_time, end_time)

# Display the generated sales report visualization
sales_report_fig.show()
```

# **ðŸ”Ingredients Consumption report**

```python
# Function to generate report with respect ingredients used

def sales_report_by_ingredient_visualization(start_date, end_date, months, start_time=None, end_time=None):
    # Filter sales data based on provided dates and months
    filtered_df = df[(df['Time'] >= start_date) & (df['Time'] <= end_date) & (df['Time'].dt.month.isin(months))]

    # Further filter data based on provided time range, if given
    if start_time and end_time:
        filtered_df = filtered_df[(filtered_df['Time'].dt.time >= start_time) & (filtered_df['Time'].dt.time <= end_time)]

    # Sum up ingredient consumption across all filtered records
    ingredient_consumption = filtered_df.drop(['Time', 'Burger Name', 'Base type', 'Quantity Sold', 'Price'], axis=1).sum()

    # Create a DataFrame for the ingredient sales report
    report = pd.DataFrame({'Ingredient': ingredient_consumption.index, 'Total Consumption': ingredient_consumption.values})

    # Create a bar chart visualization for the ingredient sales report
    fig = px.bar(report, x='Ingredient', y='Total Consumption', color='Ingredient', title='Sales Report by Ingredient',
                 labels={'Total Consumption': 'Total Consumption'}, text='Total Consumption')

    # Customize the appearance of the bar chart
    fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
    fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')

    return fig
```

## **ðŸ›  Filters to adjust desired range**

```python
# Example usage with time filter
start_date = datetime(2024, 1, 1)  # Start date for the report
end_date = datetime(2024, 12, 31)  # End date for the report
months = [1, 2, 3, 6, 7, 8]  # Selected months: January, February, March, June, July, August

# Define the time range for filtering
start_time = datetime.strptime('16:00:00', '%H:%M:%S').time()  # Start time for the report
end_time = datetime.strptime('18:00:00', '%H:%M:%S').time()  # End time for the report

# Generate the ingredient sales report visualization for the specified date and time range
ingredient_sales_report_fig = sales_report_by_ingredient_visualization(start_date, end_date, months, start_time, end_time)

# Display the generated ingredient sales report visualization
ingredient_sales_report_fig.show()
```

# **âš’ Slicers including BURGER NAME**

```python
# Function to create report

def sales_report_by_burger_name_visualization(start_date, end_date, start_time=None, end_time=None, burger_name=None):
    # Print selected burger name, date, and time
    print("...............................................")
    print(f"Selected Burger: {burger_name}")
    print("...............................................")
    print(f"Selected Date Range: {start_date} to {end_date}")
    print(f"Selected Time Range: {start_time} to {end_time}")
    print("...............................................")

    # Filter sales data based on provided dates
    filtered_df = df[(df['Time'] >= start_date) & (df['Time'] <= end_date)]
    if start_time and end_time:  # Further filter based on time range if provided
        filtered_df = filtered_df[(filtered_df['Time'].dt.time >= start_time) & (filtered_df['Time'].dt.time <= end_time)]

    # Convert month numbers to month names for readability
    filtered_df['Month'] = filtered_df['Time'].dt.month_name()

    # Categorize time as peak or non-peak hours
    filtered_df['Time Category'] = 'Non-Peak Hours'  # Default category
    peak_start_time = datetime.strptime('16:00:00', '%H:%M:%S').time()  # Define start of peak hours
    peak_end_time = datetime.strptime('20:30:00', '%H:%M:%S').time()  # Define end of peak hours
    filtered_df.loc[(filtered_df['Time'].dt.time >= peak_start_time) & (filtered_df['Time'].dt.time <= peak_end_time), 'Time Category'] = 'Peak Hours'

    # Group by month and time category, then sum up quantity sold
    sales_report = filtered_df.groupby(['Month', 'Time Category'])['Quantity Sold'].sum().reset_index()

    # Create visualization for quantity sold by month and time category
    fig1 = go.Figure()  # Initialize a figure
    for time_category in sales_report['Time Category'].unique():  # Loop over each time category
        time_category_data = sales_report[sales_report['Time Category'] == time_category]  # Filter data for the time category
        fig1.add_trace(go.Bar(x=time_category_data['Month'], y=time_category_data['Quantity Sold'], name=time_category))  # Add a bar trace to the figure
    fig1.update_layout(barmode='group', xaxis_title='Month', yaxis_title='Quantity Sold',
                       title='Quantity Sold by Month and Time Category')  # Update layout with titles and group bars

    # Filter sales data based on provided burger name if given
    if burger_name:
        filtered_df = filtered_df[filtered_df['Burger Name'] == burger_name]

    # Sum up ingredient consumption
    ingredient_consumption = filtered_df.drop(['Time', 'Burger Name', 'Base type', 'Quantity Sold', 'Price'], axis=1).sum()
    ingredient_report = pd.DataFrame({'Ingredient': ingredient_consumption.index, 'Total Consumption': ingredient_consumption.values})

    # Create visualization for ingredient consumption
    fig2 = go.Figure()  # Initialize a figure
    fig2.add_trace(go.Bar(x=ingredient_report['Ingredient'], y=ingredient_report['Total Consumption'], marker=dict(color='blue'), name='Ingredient Consumption'))  # Add a bar trace
    fig2.update_layout(title=f'Ingredient Consumption for {burger_name}', xaxis_title='Ingredient', yaxis_title='Total Consumption')  # Update layout with titles

    # Calculate KPIs (Key Performance Indicators)
    total_revenue = df['Price'].sum()  # Calculate total revenue
    total_quantity_sold = df['Quantity Sold'].sum()  # Calculate total quantity sold
    average_revenue_per_burger = df['Price'].mean()  # Calculate average revenue per burger
    peak_hour_sales = df[(df['Time'].dt.time >= peak_start_time) & (df['Time'].dt.time <= peak_end_time)]['Quantity Sold'].sum()  # Calculate peak hour sales
    plant_based_sales = df[df['Base type'] == 'Plant-Based']['Quantity Sold'].sum()  # Calculate plant-based burger sales
    beef_based_sales = df[df['Base type'] == 'Beef-Based']['Quantity Sold'].sum()  # Calculate beef-based burger sales
    most_popular_burger = df.groupby('Burger Name')['Quantity Sold'].sum().idxmax()  # Determine the most popular burger
    total_ingredient_consumption = df.drop(['Time', 'Burger Name', 'Base type', 'Quantity Sold', 'Price'], axis=1).sum().sort_values(ascending=False)  # Calculate total ingredient consumption
    df['Month'] = df['Time'].dt.month  # Extract month from time for monthly analysis
    average_sales_per_month = df.groupby('Month')['Quantity Sold'].mean()  # Calculate average sales per month

    # Print KPI details
    print("\nKPIs (-:Key point indicatiors:-)")
    print("====================================================================")
    print(f"::Total Revenue: ${total_revenue}")
    print("-----------------------------------")
    print(f"::Total Quantity Sold: {total_quantity_sold}")
    print("-----------------------------------")
    print(f"::Average Revenue per Burger: ${average_revenue_per_burger}")
    print("-----------------------------------")
    print(f"::Peak Hour Sales: {peak_hour_sales}")
    print("-----------------------------------")
    print(f"::Plant-Based Sales: {plant_based_sales}")
    print("-----------------------------------")
    print(f"::Beef-Based Sales: {beef_based_sales}")
    print("-----------------------------------")
    print(f"::Most Popular Burger: {most_popular_burger}")
    print("-----------------------------------")
    print(f"\n-:Total Ingredient Consumption:-\n**********************************\n{total_ingredient_consumption}")
    print("-----------------------------------")
    print(f"\n-:Average Sales per Month:-\n{average_sales_per_month}")
    print("**********************************")
    print("=====================================")

    return fig1, fig2  # Return the two generated figures
```
